### Dependancies

```python
import pandas as pd
from functools import reduce
```

### Importing Data Sets


```python
cb_jun19 = pd.read_csv('../CitiBike_DataSets/JC-201906-citibike-tripdata.csv')
cb_jul19 = pd.read_csv('../CitiBike_DataSets/JC-201907-citibike-tripdata.csv')
cb_aug19 = pd.read_csv('../CitiBike_DataSets/JC-201908-citibike-tripdata.csv')
cb_sept19 = pd.read_csv('../CitiBike_DataSets/JC-201909-citibike-tripdata.csv')
cb_oct19 = pd.read_csv('../CitiBike_DataSets/JC-201910-citibike-tripdata.csv')
cb_nov19 = pd.read_csv('../CitiBike_DataSets/JC-201911-citibike-tripdata.csv')
cb_dec19 = pd.read_csv('../CitiBike_DataSets/JC-201912-citibike-tripdata.csv')
cb_jan20 = pd.read_csv('../CitiBike_DataSets/JC-202001-citibike-tripdata.csv')
cb_feb20 = pd.read_csv('../CitiBike_DataSets/JC-202002-citibike-tripdata.csv')
cb_mar20 = pd.read_csv('../CitiBike_DataSets/JC-202003-citibike-tripdata.csv')
cb_apr20 = pd.read_csv('../CitiBike_DataSets/JC-202004-citibike-tripdata.csv')
cb_may20 = pd.read_csv('../CitiBike_DataSets/JC-202005-citibike-tripdata.csv')
cb_jun20 = pd.read_csv('../CitiBike_DataSets/JC-202006-citibike-tripdata.csv')
```

#### Data Reading

#### Cleaning Data


```python
citibike_data = [cb_jun19, cb_jul19, cb_aug19, 
                    cb_sept19, cb_oct19, cb_nov19, 
                    cb_dec19, cb_jan20, cb_feb20, 
                    cb_mar20, cb_apr20, cb_may20, cb_jun20]

citibike_merge = reduce(lambda left,right: pd.merge(left, right, how="outer"), citibike_data)
```

```python
# Cleaning Gender Column
citibike_merge["gender"].replace({0: "unknown", 1: "male", 2: "female"}, inplace=True)
citibike_merge.head()
```
